```python
attacks_entero = pd.read_csv("attacks_limpieza_completa.csv", index_col= 0)

#Cogemos las primeras columnas para seguir el enunciado y no tener las variables categóricas codificadas:

attacks_cols = attacks_entero[['year', 'type', 'country', 'age', 'species_', 'fecha_limpia', 'fatal',
       'sex', 'latitud', 'longitud']]

attacks_cols.head(2)
```

```python
#Sacamos las columnas con las que vamos a trabajar, las de los 5 países con más ataques:

attacks = attacks_cols[attacks_cols["country"].isin(["usa", "australia", "new zealand", "south africa", "papua new guinea"])]
attacks
```

```python
clima = pd.read_csv("datos_clima.csv", index_col = 0).reset_index()
clima.head()
```

```python
#Como en este df el country aparece en mayúsculas, reciclamos una función para transformar a minúsculas:
def minusculizator (columna):
    try:
        columna = columna.lower()
    except:
        pass
    return columna

clima["country"] = clima["country"].apply(minusculizator)
clima.head(2)
```

Desempacamos las columnas "wind_profile" y rh_profile:

```python
import ast

#WIND
clima['wind_profile'] = clima['wind_profile'].apply(ast.literal_eval)
wind = clima['wind_profile'].apply(pd.Series)
```

```python
#RH
clima['rh_profile'] = clima['rh_profile'].apply(ast.literal_eval)
rh = clima['rh_profile'].apply(pd.Series)
```

```python
for i in range(len(wind.columns)):
    #Wind: direction & speed
    nombre = "wind_dir_" + str(wind[i].apply(pd.Series)["layer"][0])
    nombre2 = "wind_speed_" + str(wind[i].apply(pd.Series)["layer"][0])
    direc = list(wind[i].apply(pd.Series)["direction"])
    speed = list(wind[i].apply(pd.Series)["speed"])
    
    #RH
    nombre_rh = "rh_" + str(rh[i].apply(pd.Series)["layer"][0])
    valor_rh = list(rh[i].apply(pd.Series)["rh"])
    

    clima.insert(i, nombre, direc)
    clima.insert(i, nombre2, speed)
    clima.insert(i, nombre_rh, valor_rh)
```

```python
clima.head(2)
```

# Medias de las columnas climáticas por país

```python
clima_medias_pais = clima.groupby("country").mean()
clima_medias_pais
```

```python
print(clima.shape)
print(attacks.shape)

attacks["country"].unique()
```


Pensar Por qué el merge nos quita filas

```python
juntos = attacks.merge(clima_medias_pais, on = "country")
juntos
```
