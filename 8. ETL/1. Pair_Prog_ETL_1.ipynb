üí° Pistas üí°:

*Crearos un diccionario donde:*

-Las keys sean los nombres de los paises

-Los values sean las coordenadas de los paises en formato lista:

Iterar por el diccionario. Dentro del for haremos la llamada a la API como hemos aprendido en las clases invertidas. Recordamos que para iterar por el diccionario tenemos que seguir la siguiente estructura:

-Crearnos un dataframe vac√≠o donde iremos concatenando los datos obtenidos de la API cada pa√≠s.

üë©üèΩ‚Äçü¶∞üë©üèΩ‚Äçü¶±üí¨   
- Creamos el diccionario con pa√≠s y longitud/latitud

```python
dicc_paises = {"USA": ["39.7837304","-100.445882"], "Australia":["-24.7761086","134.755"],"South_Africa": ["-28.8166236","24.991639"],"New_Zealand": ["-41.5000831","172.8344077"], "Papua_New_Guinea": ["-5.6816069", "144.2489081"]}
```

üë©üèΩ‚Äçü¶∞üë©üèΩ‚Äçü¶±üí¨   
- Comprobamos que los enlaces que resultan son correctos:

```python
for key, value in dicc_paises.items():
    print( f'http://www.7timer.info/bin/api.pl?lon={value[1]}&lat={value[0]}&product=meteo&output=json')
```

üë©üèΩ‚Äçü¶∞üë©üèΩ‚Äçü¶±üí¨   
- Creamos un data frame vac√≠o donde:
- a) Iteramos por cada uno de los enlaces que resultan del diccionario (pa√≠s: lat/long)
- b) Transformamos el contenido de la API para cada pa√≠s en un dataframe (cogemos el contenido del json de "dataseries" que es donde vienen los datos)
- c) A√±adimos filas a la columna "country" con el pa√≠s que toca en cada iteraci√≥n
- d) Concatenamos los dataframes resultantes

```python
df_juntos = pd.DataFrame()
print(dicc_paises.keys())

for key, value in dicc_paises.items():
    print(key)
    response = requests.get(url=f'http://www.7timer.info/bin/api.pl?lon=-{value[1]}&lat={value[0]}&product=meteo&output=json')
    print(df_juntos.shape)
    contenido = pd.DataFrame.from_dict(pd.json_normalize(response.json()['dataseries']))
    contenido["country"] = key
    df_juntos = pd.concat([df_juntos, contenido], axis = 0)
    print(df_juntos.shape)   
```

```python
df_juntos
```

üë©üèΩ‚Äçü¶∞üë©üèΩ‚Äçü¶±üí¨   
- Como necesit√°bamos crear la columna "country" despu√©s de a√±adir el "contenido" de cada pa√≠s para saber el n√∫mero de filas, despu√©s reordenamos las columnas para que "country" quede al inicio y facilitar la lectura del dataframe completo:

```python
cols = df_juntos.columns 
new_cols = ['country', 'timepoint', 'cloudcover', 'highcloud', 'midcloud', 'lowcloud',
       'rh_profile', 'wind_profile', 'temp2m', 'lifted_index', 'rh2m',
       'msl_pressure', 'prec_type', 'prec_amount', 'snow_depth',
       'wind10m.direction', 'wind10m.speed', ]
df_juntos = df_juntos.reindex(columns = new_cols)
```
